```python
import pandas as pd
from sklearn.linear_model import LinearRegression
import numpy as np

df = pd.read_csv("/content/drive/My Drive/Colab Notebooks/rainfaLLIndia.csv")
display(df)
```

```python
# sd0 = df.iloc[:19,:]  #tested whole prediction code by taking only 20 rows of andman&nicobar   ----working

all_subdivisons = df[df.duplicated(subset='subdivision', keep=False)]['subdivision'].unique()
# print(all_subdivisons)


all_subdivisons_list = []
for i in all_subdivisons:
  all_subdivisons_list.append(i)

# print(all_subdivisons_list)
# print(all_subdivisons_list[6])

for s in all_subdivisons_list :
 sd = df[df['subdivision']== s]
 # print(sd)

 df1 = sd.drop('subdivision', axis=1)
 # print(df1)

 x = df1.iloc[:,:1].values
 # print(x)

 y = df1.iloc[:,-1:].values
 # print(y)

 model = LinearRegression()
 model.fit(x,y)

 inYear = df['YEAR'].max()+1

#  subdivision = 'ANDAMAN & NICOBAR ISLAND'
 predicted_rainfall = model.predict([[inYear]])
 # print(f'Predicted Rainfall (Jun-Sep) for the year 2022: {predicted_rainfall[0]} mm')

 if predicted_rainfall>=12000.99:
   print(f'Highest Rainfall Predicted in {s} for the year {inYear} (June-Sept): upto approx. {predicted_rainfall[0]} mm')
 elif predicted_rainfall>=5000.99:
   print(f'Maximum Rainfall Predicted in {s} for the year {inYear} (June-Sept): upto approx. {predicted_rainfall[0]} mm')
 elif predicted_rainfall>=1000.99 or predicted_rainfall>=400.99:
   print(f' \033[1m Average \033[0m Rainfall Predicted in \033[1m{s}\033[0m for the year \033[1m{inYear}\033[0m (June-Sept): upto approx. {predicted_rainfall[0]} mm')
 elif predicted_rainfall>=250.99:
   print(f'Minimum Rainfall Predicted in {s} for the year {inYear} (June-Sept): upto approx. {predicted_rainfall[0]} mm')
 else :
   print(f'Drought Predicted in {s} for the year {inYear} (June-Sept): upto approx. {predicted_rainfall[0]} mm')

```
